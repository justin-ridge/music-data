```python
import pandas as pd
import numpy as np
df = pd.read_csv('spotify_features.csv')
```

```python
df.head()
```

```python
df = df[df['genre']=='Pop']
```

```python
df.head()
```

```python
df.describe()
```

```python
df['hit'] = df['popularity'] >= 75
```

```python
df.describe()
```

```python
df.head()
```

```python
df = df.drop(['genre','artist_name','track_name'],axis=1)
```

```python
df.head()
```

```python
df.to_csv('pop.csv')
```

```python
features_transformed = df.drop(['hit'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(features_transformed)
features_transformed = mms.transform(features_transformed)
```

```python
df.columns
```

```python
minmax = pd.DataFrame(features_transformed,columns=['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo',
       'valence'])
```

```python
minmax.head()
```

```python
minmax.mean()
```

```python
minmax.to_csv('pop_features_minmax.csv')
```
